```python
# Function  --> Collection of statements 

# Program  ---> collection of statements  ---> collection of functions 

# process  ---> the program is in execution 

# 2 tasks at time  ---> one process is completed then it will go for 2nd task

# thread -> ( thread & process can run at time  )

# Parent process 
'''
  # thread1
  
  # thread2
  
  # threads
'''
# parents process
from time import sleep,time
print("start time : ",time())
def fun1(name= None):
    for i in range(10):
        print("name : ",name)
        sleep(2)  # 2 sec
def fun2(phone= None):
    for i in range(10):
        print("phone : ",phone)
        sleep(2)  # 2 sec    
        
fun1("anjan") # function call
print("**************")
fun2("54895485945")
print("**************")
print("stop time : ",time())
```

```python
# Python program to illustrate the concept 
# of threading 
import threading 
import os 
  
def task1(): 
    print("Task 1 assigned to thread: {}".format(threading.current_thread().name)) 
    print("ID of process running task 1: {}".format(os.getpid())) 
  
def task2(): 
    print("Task 2 assigned to thread: {}".format(threading.current_thread().name)) 
    print("ID of process running task 2: {}".format(os.getpid())) 
  
if __name__ == "__main__": 
  
    # print ID of current process 
    print("ID of process running main program: {}".format(os.getpid())) 
  
    # print name of main thread 
    print("Main thread name: {}".format(threading.current_thread().name)) 
  
    # creating threads 
    t1 = threading.Thread(target=task1, name='t1') 
    t2 = threading.Thread(target=task2, name='t2')   
  
    # starting threads 
    t1.start() 
    t2.start() 
  
    # wait until all threads finish 
    t1.join() 
```

```python
import threading
lock = threading.Lock()

def first_function():
    for i in range(5):
        lock.acquire()  # locking 
        print ('1st lock acquired')
        print ('Executing the first funcion')
        lock.release() # releasing
        #print("1st unlock")

def second_function():
    for i in range(5):
        lock.acquire()
        print ('2nd lock acquired')
        print ('Executing the second funcion')
        lock.release()
        print("2nd lock unlock")

if __name__=="__main__":
    thread_one = threading.Thread(target=first_function)
    thread_two = threading.Thread(target=second_function)

    thread_one.start()
    thread_two.start()

    thread_one.join()
    thread_two.join()
```
